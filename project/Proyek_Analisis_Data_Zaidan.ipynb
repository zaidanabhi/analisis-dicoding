### Eksplorasi 'hr' dalam hour_df

Selanjutnya, kita akan memeriksa kolom 'hr' pada dataframe hour_df 

```python
hour_df['hr']
```

```python
hour_df['hr'].describe()
```

Bisa dilihat 'hr' mengandung nilai yang berentang dari 0-23 yang mempresentasikan waktu jam dari hari.
Namun, kita ingin menjawab pertanyaan pada "rentang jam" berapa orang-orang menggunakan layanan Bike Sharing, bukan pada suatu jam tertentu.\
Karena itu kita bisa mengelompokkan jam-jam tersebut menjadi beberapa kelompok. Sistem pengelompokkan yang saya pratikkan adalah sebagai berikut :

Grup 1 : jam 0 - 6\
Grup 2 : jam 6 - 12\
Grup 3 : jam 12 - 18\
Grup 4 : jam 18 - 0

```python
grup1 = list(range(0, 7))
grup2 = list(range(6, 13))
grup3 = list(range(12, 19))
grup4 = list(range(18, 24)) + [0]

def assign_group(hour):
    if hour in grup1:
        return 'Midnight to Morning'
    elif hour in grup2:
        return 'Morning to Noon'
    elif hour in grup3:
        return 'Noon to Evening'
    elif hour in grup4:
        return 'Evening to Midnight'
```

Untuk melakukan hal ini, saya membuat empat variabel yang setiapnya mengandung list angka yang sesuai dengan sistem pengelompokkan yang saya buat. Lalu saya membuat sebuah function yang akan menugaskan setiap variabel-variabel tersebut kepada title yang sesuai seperti berikut :

- Grup 1 : Midnight to Morning
- Grup 2 : Morning to Noon
- Grup 3 : Noon to Evening
- Grup 4 : Evening to Midnight

```python
hour_df['group_hour'] = hour_df['hr'].apply(assign_group)

# Group by the 'group' column and calculate the mean of the 'count' column
grouped_data = hour_df.groupby('group_hour')['cnt'].mean()

group_order = ['Midnight to Morning', 'Morning to Noon', 'Noon to Evening', 'Evening to Midnight']
grouped_data_sorted = grouped_data.reindex(group_order)

print(grouped_data_sorted)
```

Lalu saya menugaskan function tersebut kepada variabel hour_df['group_hour'] yang akan menghasilkan kolom bernama group_hour. Karena kolom 'group_hour' merupakan hasil pengelompokkan dari nilai pada kolom 'hr' kita bisa memeriksa hasil nilai mean kolom 'cnt' untuk setiap kelompok rentang jam. Karena kita masih pada proses eksplorasi, kita hanya memeriksa nilai mean pada kolom 'cnt'. Pada section Explanatory Analysis, saya akan menunjukkan hasil nilai mean pada kolom pengunjung lain untuk mendapatkan jawaban yang lebih pasti.

## Visualization & Explanatory Analysis

### Pertanyaan 1: Pada Musim apa orang-orang sering menggunakan layanan Bike Sharing?

```python
split_by_season_daypeople = day_df.groupby('season')[['casual', 'registered']].mean()
split_by_season_daypeople = split_by_season_daypeople.rename(index=season_names)
print(split_by_season_daypeople)
```

```python
cnt_by_season_daypeople = day_df.groupby('season')['cnt'].mean()
cnt_by_season_daypeople = cnt_by_season_daypeople.rename(index=season_names)
print(cnt_by_season_daypeople)
```

Kita periksa jumlah orang secara rata-rata pada keempat musim pertama per hari. Tabel pertama untuk nilai rata-rata dari user casual dan registered, dan tabel kedua untuk nilai rata-rata user secara keseluruhan.

```python
split_by_season_hourpeople = hour_df.groupby('season')[['casual', 'registered']].mean()
split_by_season_hourpeople = split_by_season_hourpeople.rename(index=season_names)
print(split_by_season_hourpeople)
```

```python
cnt_by_season_hourpeople = hour_df.groupby('season')['cnt'].mean()
cnt_by_season_hourpeople = cnt_by_season_hourpeople.rename(index=season_names)
print(cnt_by_season_hourpeople)
```

Lalu kita periksa jumlah orang secara rata-rata pada keempat musim pertama per jam. Bisa dilihat hasil nilai mean dilihar dari day_df dan hour_df kelihatan berbeda namun jika kita bandingkan kedua tabel, kedua dataframe menghasilkan kesimpulan yang sama, yaitu : Orang-orang lebih banyak menggunakan layanan Bike Sharing pada Musim Panas (Summer).

```python
split_by_season_daypeople_df = split_by_season_daypeople.reset_index()

mean_by_season_melted_day = split_by_season_daypeople_df.melt(id_vars='season', var_name='user_type', value_name='mean_value')

plt.figure(figsize=(10, 6))
sns.barplot(data=mean_by_season_melted_day, x='season', y='mean_value', hue='user_type')
plt.title('Casual and Registered Users by Season (counted per day)')
plt.xlabel('Season')
plt.ylabel('Mean Value')
plt.legend(title='User Type')
plt.show()
```

Lalu kita visualisasikan hasil dari temuan kita. Pertama kita visualisasikan nilai mean dari casual user dan registered user. Karena analisa ini menggunakan dua variabel, kita gunakan Clustered Bar Chart.

```python
cnt_by_season_daypeople.plot(kind='pie', autopct='%1.1f%%', startangle=180)
plt.ylabel('')
plt.title('Bike Sharing Users Percentage by Season (counted per day)')
plt.show()
```

Setelah itu divisualisasikan persentase dari total count user setiap Musim. Analisa ini hanya menggunakan satu variabel jadi kita bisa menggunakan Pie Chart.

```python
split_by_season_hourpeople_df = split_by_season_hourpeople.reset_index()

mean_by_season_melted_hour = split_by_season_hourpeople_df.melt(id_vars='season', var_name='user_type', value_name='mean_value')

plt.figure(figsize=(10, 6))
sns.barplot(data=mean_by_season_melted_hour, x='season', y='mean_value', hue='user_type')
plt.title('Casual and Registered Users by Season (counted per hour)')
plt.xlabel('Season')
plt.ylabel('Mean Value')
plt.legend(title='User Type')
plt.show()
```

Lalu kita visualisasikan nilai mean user casual dan registered pada setiap musim per jam. Bisa dilihat Chart yang dihasilkan menunjukkan kesimpulan yang sama dengan Chart per hari.

```python
cnt_by_season_hourpeople.plot(kind='pie', autopct='%1.1f%%', startangle=180)
plt.ylabel('')
plt.title('Users Percentage by Season (counted per hour)')
plt.show()
```

Terakhir kita visualisasikan persentase dari total count user setiap Musim per jam. Bisa dilihat perbandingan persentase setiap Musim pada Chart ini mirip dengan visualisasi Bar Chart total count user per hari. Karena semua Chart telah terbukti tidak menkontradiksi satu sama lain, bisa disimpulkan bahwa Musim dimana orang-orang paling banyak menggunakan layanan Bike Sharing adalah Musim Panas (Summer).

### Pertanyaan 2: Pada rentang jam berapa orang-orang banyak menggunakan layanan Bike Sharing?

```python
print(grouped_data_sorted)
```

Untuk menjawab pertanyaan ini, kita analisa nilai mean user pada setiap rentang jam. Pengelompokkan dan pencarian nilai mean total count user sudah kita lihat pada bagian EDA. Hasil ini saya tampilkan lagi untuk memudahkan membaca analisis tanpa harus memeriksa section EDA lagi.

```python
grouped_data_split = hour_df.groupby('group_hour')[['casual', 'registered']].mean()

group_order = ['Midnight to Morning', 'Morning to Noon', 'Noon to Evening', 'Evening to Midnight']
grouped_data_sorted_split = grouped_data_split.reindex(group_order)

print(grouped_data_sorted_split)
```

Selanjutnya kita cari nilai mean pada user casual dan user registered pada setiap rentang jam. Hal ini dilakukan untuk membuktikan hasil analisa tidak kontradiksi disaat dilihat dari user casual dan registered ataupun dari total count user.

```python
grouped_data_sorted.plot(kind='pie', autopct='%1.1f%%', startangle=180)
plt.ylabel('')
plt.title('Users Percentage by Hour Range')
plt.show()
```

Setelah itu divisualisasikan hasil analisa dengan Bar Chart. Bisa dilihat pada chart kebanyakan user menggunakan layanan pada rentang jam siang sampai sore.

```python
grouped_data_sorted_split.reset_index(inplace=True)

melted_data = grouped_data_sorted_split.melt(id_vars='group_hour', var_name='user_type', value_name='mean_value')

plt.figure(figsize=(10, 6))
sns.barplot(data=melted_data, x='group_hour', y='mean_value', hue='user_type')
plt.title('Casual and Registered Users by Hour Range')
plt.xlabel('Hour Group')
plt.ylabel('Mean Value')
plt.xticks(rotation=0)
plt.legend(title='User Type')
plt.tight_layout()
plt.show()
```

Lalu kita visualisasikan hasil analisa nilai mean user casual dan registered menggunakan Clustered Bar Chart. Chart menunjukkan bahwa rentang jam dimana user casual dan registered paling banyak menggunakan layanan Bike Sharing adalah di rentang jam siang sampai sore. Hasil analisa ini sesuai dengan Pie Chart total count user, yang mengartikan bahwa telah terbukti rentang jam dimana orang-orang menggunakan layanan Bike Sharing adalah siang sampai sore (Noon to Evening).
